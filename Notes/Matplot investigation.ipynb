```python
# We need to move to a core+translations paradigm.
# Translation axes are not going to always be x and y.
def tile(polyList,rTrans,uTrans):
    ''' draw each of polyList, then tile the plane with them translating rTrans vectors right and uTrans up. ''' 
    fig = plt.figure(figsize=(5,5))
    plt.axis('off')
    ax = plt.axes()
    w,h = (4,4)
    ax.set_xlim(-w,w)
    ax.set_ylim(-h,h)
    for rStep in range(-w,w):
        for uStep in range(-h,h):
            rotter = trn.Affine2D().translate(*rStep*rTrans).translate(*uStep*uTrans) + ax.transData
            for poly in polyList:
                pColor = random.rand(3,1)
                basePoly = plt.Polygon(poly,
                                       facecolor=pColor,
                                       edgecolor=pColor)
                basePoly.set_transform(rotter)
                ax.add_patch(basePoly)
    plt.show()
```

```python
# Let's define some primitives:
class Triangle(plt.Polygon):
    def __init__(self):
        plt.Polygon.__init__(self, [[0,0],[0.5,0.8660254037844386],[1,0]])
class Square(plt.Polygon):
    def __init__(self):
        plt.Polygon.__init__(self, [[0,0],[0,1],[1,1],[1,0]])
class Hexagon(plt.Polygon):
    def __init__(self):
        plt.Polygon.__init__(self, [[-0.8660254037844386,-0.5],[-0.8660254037844386,0.5],
                       [0,1],[0.8660254037844386,0.5],[0.8660254037844386,-0.5],[0,-1]])
#TODO: make octagon, 12gon.
```

```python
def drawPolyList(listOfPolys):
    ''' Draw passed polys. '''
    # Set up graph.
    fig = plt.figure(figsize=(5,5))
    plt.axis('off')
    ax = plt.axes()
    # Add all the poly patches.
    for poly in listOfPolys:
        # Hack to keep MPL from changing the passed polys! I hate this library!
        addPoly = copy(poly)
        addPoly.set_transform(poly.get_transform() + ax.transData)
        ax.add_patch(addPoly)
    plt.show()
```
